```python
def plot_metrics(df, ymin, ymax, tick_width, metric, parameter, legend_loc):
    plt.figure(figsize=(10, 6.18))    

    ax = plt.subplot(111)    
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)    
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)    

    ax.get_xaxis().tick_bottom()    
    ax.get_yaxis().tick_left()    

    plt.xlim(1, 5)    
    plt.ylim(ymin-0.01, ymax)    

    plt.xticks(range(1, 6))

    for y in np.arange(ymin, ymax+0.1, tick_width):    
        plt.plot(range(1, 6), [y] * len(range(1, 6)), "--", lw=0.5, color="black", alpha=0.3)    


    plt.tick_params(axis="both", which="both", bottom="off", top="off",    
                    labelbottom="on", left="off", right="off", labelleft="on")    


    ax.plot(df.Epoch, df['15-9-7-3'], lw=1.5, color='#E45555')
    ax.plot(df.Epoch, df['10-7-5-3'], lw=1.5, color='#FF8D4B')
    ax.plot(df.Epoch, df['7-7-3-3'], lw=1.5, color='#4BA0FF')
    ax.plot(df.Epoch, df['5-5-3-3'], lw=1.5, color='#15BD24')
    ax.legend(loc = legend_loc, fontsize = 12)
    plt.title('{} vs {}'.format(metric, parameter), fontsize = 14, y = 1.03)
    plt.xlabel('Epoch', fontsize = 12, labelpad = 10)
    plt.ylabel('{}'.format(metric), fontsize = 12, labelpad = 10)
    plt.show()
```

```python
plot_metrics(train_acc, 0.4, 1, 0.1, 'Training Accuracy', 'Kernel Size', 4)
plot_metrics(train_loss, 0, 2.5, 0.5, 'Training Loss', 'Kernel Size', 1)
# plot_metrics(valid_acc, 0.7, 1, 0.05, 'Validation Accuracy', 'Neural Network Depth', 4)
# plot_metrics(valid_loss, 0, 0.8, 0.1, 'Validation Loss', 'Neural Network Depth', 1)
```
