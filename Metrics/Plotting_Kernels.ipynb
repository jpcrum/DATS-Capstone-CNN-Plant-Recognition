```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
weights_df = pd.read_csv("C:/Users/sjcrum/Documents/GitHub/Plant-Image-Recognition/DATS-Capstone-CNN-Plant-Recognition/Models/kernelsGreen.csv")

print(weights_df.head())
```

```python
weights = weights_df.iloc[:,1:]
kernels = []
for i in range(len(weights)):
    kernel = np.reshape(weights.iloc[i,:], (5,5))
    kernels.append(kernel)
    if i == 0:
        plt.imshow(kernel, cmap = 'gray')
        plt.show()
```

```python
fig, ax = plt.subplots(4,8, figsize = (30,20))

index = 0
for i in range(4):
    for j in range(8):
        ax[i,j].imshow(kernels[index], cmap = "summer")
        index += 1 

plt.show()
     
```
